```python
# Calculating rating classwise
questionmeans=[]
classlist=[]
questions=[]
totalplotdata=pd.DataFrame(list(zip(classlist,questions,questionmeans)),columns=['class','questions','mean'])
totalplotdata.head()
```

```python
for class_num in range(1,13):
    class_data=data[(data['class']==class_num)]
    questionmeans=[]
    classlist=[]
    questions=[]
    for num in range(1,13):
        questions.append(num)
    #Class related questions are from Q1 to Q12
    for col in range(5,17):
        questionmeans.append(class_data.iloc[:,col].mean())
    classlist = 12*[class_num]
    plot_data=pd.DataFrame(list(zip(classlist,questions,questionmeans)),columns=['class','questions','mean'])
    totalplotdata=totalplotdata.append(plot_data,ignore_index=True)
print(totalplotdata)
```

```python
plt.figure(figsize=(20, 10))
sns.pointplot(x="questions", y="mean", data=totalplotdata, hue="class")
```

```python
# Calculating rating instructor wise
questionmeans = []
inslist = []
questions = []
totalplotdata = pd.DataFrame(list(zip(inslist,questions,questionmeans))
                      ,columns=['ins','questions', 'mean'])
for ins_num in range(1,4):
    ins_data = data[(data["instr"]==ins_num)]
    questionmeans = []
    inslist = []
    questions = []
    
    for num in range(13,29):
        questions.append(num)
    
    #Instructor related questions are from Q13 to Q28
    for col in range(17,33):
        questionmeans.append(ins_data.iloc[:,col].mean())
    inslist = 16 * [ins_num] 
    plotdata = pd.DataFrame(list(zip(inslist,questions,questionmeans))
                      ,columns=['ins','questions', 'mean'])
    totalplotdata = totalplotdata.append(plotdata, ignore_index=True)
print(totalplotdata)
```

```python
plt.figure(figsize = (20, 10))
sns.pointplot('questions', 'mean', hue = 'ins', data = totalplotdata)
```

```python
# Calculating which classes were not happy with Instructor 3
dataset_inst3 = data[data['instr'] == 3]
class_array_for_inst3 = dataset_inst3['class'].unique()
questions = []
classlist = []
questionsmean = []
totalplotdata = pd.DataFrame(list(zip(classlist, questions, questionsmean)), columns = ['class', 'questions', 'mean'])

for class_num in class_array_for_inst3:
    questions = []
    classlist = []
    questionsmean = []
    class_data = dataset_inst3[dataset_inst3['class'] == class_num]
    
    for num in range(1,13):
        questions.append(num)
    for col in range(5, 17):
        questionsmean.append(class_data.iloc[:,col].mean())
    classlist = 12*[class_num]
    
    plotdata = pd.DataFrame(list(zip(classlist, questions, questionsmean)), columns = ['class', 'questions', 'mean'])
    totalplotdata = totalplotdata.append(plotdata, ignore_index = True)
print(totalplotdata)
```

```python
plt.figure(figsize = (20, 10))
sns.pointplot(x = "questions", y = "mean", data = totalplotdata, hue = "class")
```
