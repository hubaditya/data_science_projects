```python
# Filling NA with any value as no specific knowledge as to what could be filled in

data = [train_df, test_df]

for dataset in data:
    dataset['Product_Category_2'] = dataset['Product_Category_2'].fillna(-2.0)
    dataset['Product_Category_3'] = dataset['Product_Category_3'].fillna(-2.0)
```

```python
# Converting Gender to numeric

data = [train_df, test_df]
genders = {'F': 0, 'M': 1}

for dataset in data:
    dataset['Gender'] = dataset['Gender'].map(genders)
```

```python
# Converting Age to numeric

data = [train_df, test_df]
age = {'0-17': 0, '18-25': 1, '26-35': 2, '36-45': 3, '46-50': 4, '51-55': 5, '55+': 6}

for dataset in data:
    dataset['Age'] = dataset['Age'].map(age)
```

```python
# Converting City_Category to numeric

data = [train_df, test_df]
cities = {'A': 0, 'B': 1, 'C': 2}

for dataset in data:
    dataset['City_Category'] = dataset['City_Category'].map(cities)
```

```python
# Converting Stay_In_Current_City_Years to numeric

data = [train_df, test_df]
no_of_years = {'0': 0, '1': 1, '2': 2, '3': 3, '4+': 4}

for dataset in data:
    dataset['Stay_In_Current_City_Years'] = dataset['Stay_In_Current_City_Years'].map(no_of_years)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error

target = 'Purchase'
predictors = train_df.columns.drop(['User_ID', 'Product_ID', 'Purchase'])
IDcol = ['User_ID','Product_ID']

def modelfit(alg, dtrain, dtest, predictors, target, IDcol, filename):
    
    alg.fit(dtrain[predictors], dtrain[target])
    dtrain_predictions = alg.predict(dtrain[predictors])

    cv_score = cross_val_score(alg, dtrain[predictors],(dtrain[target]) , cv = 20, scoring='neg_mean_squared_error')
    cv_score = np.sqrt(np.abs(cv_score))
    
    print("\nModel Report")
    print("RMSE : %.4g" % np.sqrt(mean_squared_error((dtrain[target]).values, dtrain_predictions)))
    print("CV Score : Mean - %.4g | Std - %.4g | Min - %.4g | Max - %.4g" % (np.mean(cv_score),np.std(cv_score),np.min(cv_score),np.max(cv_score)))
    
    dtest[target] = alg.predict(dtest[predictors])
    IDcol.append(target)
    submission = pd.DataFrame({ x: dtest[x] for x in IDcol})
    submission.to_csv(filename, index=False)
```

```python
def getCountVar(compute_df, count_df, var_name):
    # Get count of every variable name
    grouped_df = count_df.groupby(var_name)
    count_dict = {}
    for name, group in grouped_df:
        count_dict[name] = group.shape[0]

    # Map every row to its corresponding count value
    count_list = []
    for index, row in compute_df.iterrows():
        name = row[var_name]
        count_list.append(count_dict.get(name, 0))

    return count_list

def getPurchaseVar(compute_df, purchase_df, var_name):
    grouped_df = purchase_df.groupby(var_name)
    min_dict = {}
    max_dict = {}
    mean_dict = {}
    for name, group in grouped_df:
        min_dict[name] = min(np.array(group["Purchase"]))
        max_dict[name] = max(np.array(group["Purchase"]))
        mean_dict[name] = np.mean(np.array(group["Purchase"]))

    min_list = []
    max_list = []
    mean_list = []
    for index, row in compute_df.iterrows():
        name = row[var_name]
        min_list.append(min_dict.get(name,0))
        max_list.append(max_dict.get(name,0))
        mean_list.append(mean_dict.get(name,0))

    return min_list, max_list, mean_list
```

```python
data = [train_df, test_df]

for dataset in data:
    
    dataset["User_ID_Count"] = getCountVar(dataset, dataset, "User_ID")
    dataset['Product_ID_Count'] = getCountVar(dataset, dataset, 'Product_ID')
    dataset['Age_Count'] = getCountVar(dataset, dataset, 'Age')
    dataset['Occupation_Count'] = getCountVar(dataset, dataset, 'Occupation')
    dataset['Product_Category_1_Count'] = getCountVar(dataset, dataset,'Product_Category_1')
    dataset['Product_Category_2_Count'] = getCountVar(dataset, dataset, 'Product_Category_2')
    dataset['Product_Category_3_Count'] = getCountVar(dataset, dataset,'Product_Category_3')
    
    min_price_list, max_price_list, mean_price_list = getPurchaseVar(dataset, train_df, "User_ID")
    dataset["User_ID_MinPrice"] = min_price_list
    dataset["User_ID_MaxPrice"] = max_price_list
    dataset["User_ID_MeanPrice"] = mean_price_list
    min_price_list, max_price_list, mean_price_list = getPurchaseVar(dataset, train_df, "Product_ID")
    dataset["Product_ID_MinPrice"] = min_price_list
    dataset["Product_ID_MaxPrice"] = max_price_list
    dataset["Product_ID_MeanPrice"] = mean_price_list
```

```python
# Label Encoding User_ID and Product_ID

from sklearn.preprocessing import LabelEncoder

cat_columns_list = ["User_ID", "Product_ID"]
for var in cat_columns_list:
    lb = LabelEncoder()
    full_var_data = pd.concat((train_df[var],test_df[var]),axis=0).astype('str')
    lb.fit_transform(np.array(full_var_data))
    train_df[var] = lb.transform(np.array(train_df[var]).astype('str'))
    test_df[var] = lb.transform(np.array(test_df[var]).astype('str'))
```

```python
train_df.drop(["Purchase"], axis = 1)

# Final output to contain test_user_id, test_product_id and Purchase
test_user_id = np.array(test_df["User_ID"])
test_product_id = np.array(test_df["Product_ID"])
```

```python
import xgboost as xgb

def runXGB(train_X, train_y, test_X):
    params = {}
    params["objective"] = "reg:linear"
    params["eta"] = 0.05
    params["min_child_weight"] = 10
    params["subsample"] = 0.8
    params["colsample_bytree"] = 0.7
    params["silent"] = 1
    params["max_depth"] = 10
    #params["max_delta_step"]=2
    params["seed"] = 0
    #params['eval_metric'] = "auc"
    plst = list(params.items())
    num_rounds = 750

    xgtrain = xgb.DMatrix(train_X, label = train_y)
    model = xgb.train(plst, xgtrain, num_rounds)
    
    xgtest = xgb.DMatrix(train_X)
    pred_train_y = model.predict(xgtest)
    
    xgtest = xgb.DMatrix(test_X)
    pred_test_y = model.predict(xgtest)
    
    return pred_train_y, pred_test_y
```

```python
train_y = np.array(train_df["Purchase"])
train_X = np.array(train_df).astype('float')
test_X = np.array(test_df).astype('float')

pred_train_y, pred_test_y = runXGB(train_X, train_y, test_X)
```

```python
from sklearn.metrics import mean_squared_error

train_pred_df = pd.DataFrame({'Actual':train_y, 'Predicted':pred_train_y})
print("RMSE : %.4g" % np.sqrt(mean_squared_error((train_pred_df['Actual']).values, (train_pred_df['Predicted']).values)))
train_pred_df.to_csv("Predicted for Train.csv", index = False)
```
